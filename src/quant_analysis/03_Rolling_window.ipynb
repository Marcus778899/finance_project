# Moving Average
### $$MA=\frac{P_t+P_{t-1}+P_{t-2}+...+P_{t-n}}{n+1}$$

```python
def MA(price: pd.Series, days: int):
    ma=price.rolling(days).mean()
    ma.dropna(inplace=True)
    return ma
ma_5 = MA(df, 5)
ma_20 = MA(df, 20)
ma_5.plot(label='5-day MA',color="blue")
ma_20.plot(label='20-day MA',color="red")
```
