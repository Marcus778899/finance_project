# 特定期間總報酬

```python
start_time = '2023-01-01'
end_time = '2023-12-31'
df_slice = df.loc[start_time:end_time]
```

```python
TR = df_slice.iloc[-1]/df_slice.iloc[0]-1
TR
```

# 年化報酬率
> $$(1+r_1)(1+r_2)...(1+r_N)=1+TR$$
> $$(1+IRR)^N=1+TR$$
> $$IRR=(1+TR)^\frac{1}{N}-1$$

```python
days = len(df_slice)
IRR = (1 + TR) ** ( 252 / days ) - 1
IRR
```
